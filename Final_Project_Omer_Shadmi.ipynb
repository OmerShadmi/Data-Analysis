# 7. Detection and Treatment of Outliers

We remember those very small values we have seen before, we are gonna make sure we get rid of them (and values like them)

**First step - detection:**

First, we will check which columns we suspect to have outliers. In order to do that, we will use "describe" again.

```python
# describing the dataframe
df.describe()
```

We look for columns with anomalies in their data.

1. **'Bee_1 Pollination Activity':** The mean of the column is 0.389 and the maximum value is 18.43. That's for sure an outlier.

2. **'Maximum Upper Bloom Temperature':** The mean of the column is 82.147 and the minimum value is -100 (and we remember it also contains -86 value). They are for sure outliers.

Now that we know we have at least some outliers, we will deal with them.

We have 2 common methods to tackle this problem; z-score and IQR.

The z-score is good option when the data is normally distributed, which isn't the case here. Therfore, we will use the IQR method.

In this method the outlier data points are the ones falling below Q1–1.5 IQR or above Q3 + 1.5 IQR. The Q1 is the 25th percentile and Q3 is the 75th percentile of the dataset, and IQR represents the interquartile range calculated by Q3 minus Q1 (Q3–Q1).

```python
#create a function to find outliers using IQR

def find_outliers_IQR(df):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   return outliers
```

Let's see for example the outliers of 'Maximum Upper Bloom Temperature'.

```python
# Finding the outliers of the column 'Maximum Upper Bloom Temperature'
outliers = find_outliers_IQR(df['Maximum Upper Bloom Temperature'])

print("number of outliers: "+ str(len(outliers)))

print("max outlier value: "+ str(outliers.max()))

print("min outlier value: "+ str(outliers.min()))

print("The proportion of outliers: "+ str(len(outliers)/15289))
```

```python
for column in df.columns:

  outliers = find_outliers_IQR(df[column])

  print(column + ':')

  print("number of outliers: "+ str(len(outliers)))

  print("max outlier value: "+ str(outliers.max()))

  print("min outlier value: "+ str(outliers.min()))

  print("The proportion of outliers: "+ str(len(outliers)/15289))

  print('-------------')
```

Luckily, we have a relatively small amount of outliers, except the column 'Total Rainy Days'.

Now, after identifying them, we need to decide what to do with them.

We will use 3 different methods, for 3 different cases.

**First method - manually imputation (column 'Maximum Upper Bloom Temperature'):**

In the column 'Maximum Upper Bloom Temperature', it seems that the outliers are typos, that is, human mistake (probably just adding minus mistakenly). Therefore, we will impute this values to be the absolute value of the original value; -86 -> 86, -100 -> 100.

```python
# change the value in the column 'Maximum Upper Bloom Temperature' in the row 11555 to 86, and in row 11556 to 100

df.loc[[11555, 11556], 'Maximum Upper Bloom Temperature'] = [86, 100]
```

**Second method - removing the rows with outliers:**

This method is usefull when we have small amount of outliers, and then we don't lose a lot of data. So we will use this method on the columns that the proportion of outliers is very small (all the columns except 'Total Rainy Days').

```python
# function that will return only the not-outliers values, and the outliers will become NaN
def drop_outliers_IQR(df):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   not_outliers = df[~((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   return not_outliers
```

```python
# we will do it for each column
for column in df.columns:
  if column != 'Total Rainy Days':

    df[column] = drop_outliers_IQR(df[column])
```

```python
# now drop all the rows that contain NaN values, and reset the index

df = df.dropna().reset_index()
```

```python
# remove the column 'id' and rename the column 'index' as 'id'

df = df.drop('id', axis=1)
df = df.rename(columns={'index': 'id'})
```

```python
df.shape
```

As can see that we now have less rows than before, because we removed those with outliers of type 2.

**Third method - capping the outliers (column 'Total Rainy Days'):**

In this column, we have relatively many outliers, but all of them are outliers 'from below', that is, a lot smaller than most of the data. Therefore, it makes sense to use the capping method. In this method we set a min cap (lower limit), and if an outlier is under this cap - it will be equal to this cap.

For this lower limit, we will calculate it as the mean minus the standard deviations.

```python
# calculating the min cap
lower_limit = df['Total Rainy Days'].mean() - df['Total Rainy Days'].std()

print(lower_limit)
```

```python
# in the column 'Total Rainy Days', any value that is lower than the lower_limit - change it to the lower limit

df.loc[df['Total Rainy Days'] < lower_limit, 'Total Rainy Days'] = lower_limit
```

In conclusion, we have dealt with the outliers of the data.

To do it we will use 'MinMaxScaler' method.

This method scales each column independently, by the next formula:
